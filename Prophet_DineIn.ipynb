```python
#Importing libraries
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
#importing dataset
DataSet = pd.read_csv('Dataset.csv')
DataSet.head()
```

```python
#creating new dataframe
DineInAmount_Temp = DataSet
#changing date column name to ds and dine in amount to y
DineInAmount_Temp.rename(columns={'Date': 'ds'}, inplace=True)
DineInAmount_Temp.rename(columns={'Dine In Amount': 'y'}, inplace=True)
#fechting top five values
DineInAmount_Temp.head()
```

```python
#creating new dataframe for dine in sales
Data_DineInAmount = DineInAmount_Temp[['ds', 'y']]
Data_DineInAmount.head()
```

```python
#forecasting dine in sales
DineInAmount_m = Prophet()
DineInAmount_m.fit(Data_DineInAmount)
```

```python
#fetching bottom five predicated date rows
DineInAmount_future = DineInAmount_m.make_future_dataframe(periods=365)
DineInAmount_future.tail()
```

```python
#fetching bottom five predicated rows
DineInAmount_forecast = DineInAmount_m.predict(DineInAmount_future)
DineInAmount_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plotting dine sales with actual vs predicated
plot_plotly(DineInAmount_m, DineInAmount_forecast)
```

```python
#plotting diffrent components
plot_components_plotly(DineInAmount_m, DineInAmount_forecast)
```
