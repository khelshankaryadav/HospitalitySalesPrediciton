```python
#Importing libraries
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
#importing dataset
DataSet = pd.read_csv('Dataset.csv')
#fetching top five rows
DataSet.head()
```

```python
#creating new dataframe
Room_Temp=DataSet
#changing date column name to ds and room amount to y
Room_Temp.rename(columns={'Date': 'ds'}, inplace=True)
Room_Temp.rename(columns={'Room Amount': 'y'}, inplace=True)
#fechting top five values
Room_Temp.head()
```

```python
#creating new dataframe for room sales
Data_Room = Room_Temp[['ds', 'y']]
Data_Room.head()
```

```python
#forecasting room sales
Room_m = Prophet()
Room_m.fit(Data_Room)
```

```python
#fetching bottom five predicated date rows
Room_future = Room_m.make_future_dataframe(periods=365)
Room_future.tail()
```

```python
#fetching bottom five predicated rows
Room_forecast = Room_m.predict(Room_future)
Room_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plotting room sales with actual vs predicated
plot_plotly(Room_m,Room_forecast)
```

```python
#plotting diffrent components
plot_components_plotly(Room_m,Room_forecast)
```
