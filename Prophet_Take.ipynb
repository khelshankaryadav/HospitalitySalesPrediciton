```python
#Importing libraries
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
#importing dataset
DataSet = pd.read_csv('Dataset.csv')
#fetching top five rows
DataSet.head()
```

```python
#creating new dataframe
Dine_Temp = DataSet
#changing date column name to ds and takeaway amount to y
Dine_Temp.rename(columns={'Date': 'ds'}, inplace=True)
Dine_Temp.rename(columns={'Takeaway Amount': 'y'}, inplace=True)
#fechting top five values
Dine_Temp.head()
```

```python
#creating new dataframe for takeaway sales
Data_Dine = Dine_Temp[['ds', 'y']]
Data_Dine.head()
```

```python
#forecasting bar sales
Dine_m = Prophet()
Dine_m.fit(Data_Dine)
```

```python
#fetching bottom five predicated date rows
Dine_future = Dine_m.make_future_dataframe(periods=365)
Dine_future.tail()
```

```python
#fetching bottom five predicated rows
Dine_forecast = Dine_m.predict(Dine_future)
Dine_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plotting bar sales with actual vs predicated
plot_plotly(Dine_m, Dine_forecast)
```

```python
#plotting diffrent components
plot_components_plotly(Dine_m, Dine_forecast)
```
