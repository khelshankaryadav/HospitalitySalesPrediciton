```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
DataSet = pd.read_csv('Dataset.csv')
DataSet.head()
```

```python
BarAmount_Temp = DataSet
BarAmount_Temp.rename(columns={'Date': 'ds'}, inplace=True)
BarAmount_Temp.rename(columns={'Bar Amount': 'y'}, inplace=True)
BarAmount_Temp.head()
```

```python
Data_BarAmount = BarAmount_Temp[['ds', 'y']]
Data_BarAmount.head()
```

```python
BarAmount_m = Prophet()
BarAmount_m.fit(Data_BarAmount)
```

```python
BarAmount_future = BarAmount_m.make_future_dataframe(periods=365)
BarAmount_future.tail()
```

```python
BarAmount_forecast = BarAmount_m.predict(BarAmount_future)
BarAmount_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
plot_plotly(BarAmount_m, BarAmount_forecast)
```

```python
plot_components_plotly(BarAmount_m, BarAmount_forecast)
```
