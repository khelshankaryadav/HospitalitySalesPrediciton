```python
#Importing libraries
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
#importing dataset
DataSet = pd.read_csv('Dataset.csv')
#fetching top five rows
DataSet.head()
```

```python
#creating new dataframe
BarAmount_Temp = DataSet
#changing date column name to ds and bar amount to y
BarAmount_Temp.rename(columns={'Date': 'ds'}, inplace=True)
BarAmount_Temp.rename(columns={'Bar Amount': 'y'}, inplace=True)
#fechting top five values
BarAmount_Temp.head()
```

```python
#creating new dataframe for bar sales
Data_BarAmount = BarAmount_Temp[['ds', 'y']]
Data_BarAmount.head()
```

```python
#forecasting bar sales 
BarAmount_m = Prophet()
BarAmount_m.fit(Data_BarAmount)
```

```python
#fetching bottom five predicated date rows
BarAmount_future = BarAmount_m.make_future_dataframe(periods=365)
BarAmount_future.tail()
```

```python
#fetching bottom five predicated rows
BarAmount_forecast = BarAmount_m.predict(BarAmount_future)
BarAmount_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plotting bar sales with actual vs predicated
plot_plotly(BarAmount_m, BarAmount_forecast)
```

```python
#plotting diffrent components
plot_components_plotly(BarAmount_m, BarAmount_forecast)
```
